```python
from tkinter import *
 
# définition des gestionnaires
# d'événements :
 
def move():
    "déplacement de la balle"
    global x1, y1, dx, dy, flag
    x1, y1 = x1 +dx, y1 + dy
    if x1 >210:
        x1, dx, dy = 210, 0, 15
    if y1 >210:
        y1, dx, dy = 210, -15, 0
    if x1 <10:
        x1, dx, dy = 10, 0, -15
    if y1 <10:
        y1, dx, dy = 10, 15, 0
    can1.coords(oval1,x1,y1,x1+30,y1+30)
    if flag >0:
        fen1.after(50,move)        
 
 
def stop_it():
    "arret de l'animation"
    global flag
    flag =0
 
def start_it():
    "démarrage de l'animation"
    global flag
    if flag ==0:         
        flag =1
    move()
 

x1, y1 = 10, 10         # coordonnées initiales
dx, dy = 15, 0          # 'pas' du déplacement
flag =0                 # commutateur
 

fen1 = Tk()
fen1.title("Exercice d'animation avec Tkinter")

can1 = Canvas(fen1,bg='dark grey',height=250, width=250)
can1.pack(side=LEFT, padx =5, pady =5)
oval1 = can1.create_oval(x1, y1, x1+30, y1+30, width=2, fill='red')
bou1 = Button(fen1,text='Quitter', width =8, command=fen1.quit)
bou1.pack(side=BOTTOM)
bou2 = Button(fen1, text='Démarrer', width =8, command=start_it)
bou2.pack()
bou3 = Button(fen1, text='Arrêter', width =8, command=stop_it)
bou3.pack()

fen1.mainloop()
```

```python

import math
import tkinter as tk
 
 
def move():
    global nb_jours, xt, yt
    nb_jours += 1
    if nb_jours % 365 == 0:
        stopper()
 
    w = w_rot * nb_jours
    xt = xs + dts * math.cos(w)
    yt = ys - dts * math.sin(w)
    cadre.coords(terre, xt - rt, yt - rt, xt + rt, yt + rt)
 
    w = w_revol * nb_jours
    xm = xt - (rt + rm) * math.cos(w)
    ym = yt + (rt + rm) * math.sin(w)
   
 
    if flag:
        cadre.after(40, move)
 
 
def stopper():
    global flag
    flag = False
 
 
def demarrer():
    global flag
    flag = True
    move()
 
 
def reinitialiser():
    global flag, xt, yt, nb_jours
    nb_jours = 0
    flag = False
    xt, yt = xs + dts, ys
    cadre.coords(terre, xt - rt, yt - rt, xt + rt, yt + rt)
   
 
 
def quitter():
    root.destroy()
 
 
# SOLEIL
rs = 50                   # rayon
xs = ys = 450             # position
 
# TERRE
dts = 350                 # distance Terre - Soleil
rt = 20                   # rayon
xt, yt = xs + dts, ys     # position
rm = 3                    # rayon du repère du midi
 
nb_jours = 0
T_rot = 31556925 / (24 * 3600)
T_revol = 86164.1 / (24 * 3600)
w_rot = 2 * math.pi / T_rot
w_revol = 2 * math.pi / T_revol
 
flag = False
 
 
root = tk.Tk()
 
cadre = tk.Canvas(root, bg="dark grey", height=850, width=900)
cadre.pack(side="left")
 
soleil = cadre.create_oval(
    xs - rs,
    ys - rs,
    xs + rs,
    ys + rs,
    fill="yellow")
 
terre = cadre.create_oval(
    xt - rt,
    yt - rt,
    xt + rt,
    yt + rt,
    fill="blue")

 
tk.Button(root, text="Quitter", command=quitter).pack(side="bottom")
tk.Button(root, text="Démarrer", command=demarrer).pack()
tk.Button(root, text="Arrêter", command=stopper).pack()
tk.Button(root, text="Réinitialiser", command=reinitialiser).pack()
 
root.mainloop()
```
