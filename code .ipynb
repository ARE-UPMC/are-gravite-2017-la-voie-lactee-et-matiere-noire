```python
from tkinter import *
 
# définition des gestionnaires
# d'événements :
 
def move():
    "déplacement de la balle"
    global x1, y1, dx, dy, flag
    x1, y1 = x1 +dx, y1 + dy
    if x1 >210:
        x1, dx, dy = 210, 0, 15
    if y1 >210:
        y1, dx, dy = 210, -15, 0
    if x1 <10:
        x1, dx, dy = 10, 0, -15
    if y1 <10:
        y1, dx, dy = 10, 15, 0
    can1.coords(oval1,x1,y1,x1+30,y1+30)
    if flag >0:
        fen1.after(50,move)        
 
 
def stop_it():
    "arret de l'animation"
    global flag
    flag =0
 
def start_it():
    "démarrage de l'animation"
    global flag
    if flag ==0:         
        flag =1
    move()
 

x1, y1 = 10, 10         # coordonnées initiales
dx, dy = 15, 0          # 'pas' du déplacement
flag =0                 # commutateur
 

fen1 = Tk()
fen1.title("Exercice d'animation avec Tkinter")

can1 = Canvas(fen1,bg='dark grey',height=250, width=250)
can1.pack(side=LEFT, padx =5, pady =5)
oval1 = can1.create_oval(x1, y1, x1+30, y1+30, width=2, fill='red')
bou1 = Button(fen1,text='Quitter', width =8, command=fen1.quit)
bou1.pack(side=BOTTOM)
bou2 = Button(fen1, text='Démarrer', width =8, command=start_it)
bou2.pack()
bou3 = Button(fen1, text='Arrêter', width =8, command=stop_it)
bou3.pack()

fen1.mainloop()
```

```python
import math
from tkinter import *

#Fenetre
HEIGHT = 1100
WIDTH = 1100
Couleur_de_fond ='black'

# BULBE
rB = 100 # rayon (en dizaines de milliers d’années-lumière)
xB = WIDTH/2 # position x
yB = HEIGHT/2 # position y

# SOLEIL
distance_soleil_bulbe = 277 # distance Soleil – Bulbe (en centaines d’années-lumière)*10
rS = 7 # rayon (pas à l’échelle, rendu visible -grossi 100 millions de fois-)
xS = xB + distance_soleil_bulbe # position x
yS = yB # position y
"  " 
# PLANETE //Prototype
distance_planete_soleil = 15 + rS # distance Terre – Soleil
rP = 1 # rayon (pas à l’échelle, arbitraire)
xP = xS + distance_planete_soleil # position x
yP = yS # position y
"  "
#MERCURE
distance_mercure_soleil = 2 + rS
rM = 1
xM = xS + distance_mercure_soleil
yM = yS
c1 = 'white smoke'

#VENUS
distance_venus_soleil = 7 + rS
rV = 2
xV = xS + distance_venus_soleil
yV = yS
c2 = 'gold'

#TERRE
distance_terre_soleil = 11 + rS
rT = 2
xT = xS + distance_terre_soleil
yT = yS
c3 = 'dodger blue'

#Mars
distance_mars_soleil = 14 + rS
rMa = 1
xMa = xS + distance_mars_soleil
yMa = yS
c4 = 'orange red'

#JUPITER
distance_jupiter_soleil = 44 + rS
rJ = 6
xJ = xS + distance_jupiter_soleil
yJ = yS
c5 = 'sandy brown'

#SATURNE
distance_saturne_soleil = 60 + rS
rSa = 5
xSa = xS + distance_saturne_soleil
ySa = yS
c6 = 'bisque2'

#URANUS
distance_uranus_soleil = 80 + rS
rU = 4
xU = xS + distance_uranus_soleil
yU = yS
c7 = 'turquoise'

#NEPTUNE
distance_neptune_soleil = 100 + rS
rN = 4
xN = xS + distance_neptune_soleil
yN = yS
c8 = 'blue'

#CERCLE
rC = 120
xC = xS
yC = yS
c9 = 'white'






distance_etoile_bulbe = 175
rA = 10
xA = xB + distance_etoile_bulbe
yA = yB

#SHIFT COLOR
variable = 0
compteur = 0

#INITIALISATION 10 jours/frame pour le systeme solaire
nb_jours = 0
nb_annees = 0 #en 100aine de milliers d’années



T_rotation_soleil_bulbe = 2260
w_rot_soleil = 2 * math.pi / T_rotation_soleil_bulbe

T_rotation_etoile_bulbe = (distance_etoile_bulbe/distance_soleil_bulbe)*T_rotation_soleil_bulbe    
w_rot_etoile = 2 * math.pi / T_rotation_etoile_bulbe

T_rotation_mercure_soleil = 9
w_rot_mercure = 2 * math.pi / T_rotation_mercure_soleil

T_rotation_venus_soleil = 22
w_rot_venus = 2 * math.pi / T_rotation_venus_soleil

T_rotation_terre_soleil = 36
w_rot_terre = 2 * math.pi / T_rotation_terre_soleil

T_rotation_mars_soleil = 68
w_rot_mars = 2 * math.pi / T_rotation_mars_soleil

T_rotation_jupiter_soleil = 433
w_rot_jupiter = 2 * math.pi / T_rotation_jupiter_soleil

T_rotation_saturne_soleil = 1075
w_rot_saturne = 2 * math.pi / T_rotation_saturne_soleil

T_rotation_uranus_soleil = 3068
w_rot_uranus = 2 * math.pi / T_rotation_uranus_soleil

T_rotation_neptune_soleil = 6022
w_rot_neptune = 2 * math.pi / T_rotation_neptune_soleil

T_rotation_bulbe_soleil = 2260
w_rot_bulbe = 2 * math.pi / T_rotation_soleil_bulbe

mouvement_etat = False

MaFenetre = Tk()
MaFenetre.title('Rotations')

Fenetre = Canvas(MaFenetre, bg=Couleur_de_fond, height=HEIGHT, width=WIDTH)
Fenetre.pack(side='left')

def mouvement():
    global nb_jours, nb_annees, compteur, variable, xA, yA, xB, yB, xS, yS, xM, yM, xV, yV, xT, yT, xMa, yMa, xJ, yJ, xSa, ySa, xU, yU, xN, yN
    nb_annees += 1
    nb_jours += 1
    variable = distance(xS, xA, yS, yA)

    xB = WIDTH/2
    yB = HEIGHT/2
  
    Fenetre.coords(diametre1, xB - 500, yB, xB + 500, yB)
    Fenetre.coords(diametre2, xB, yB - 500, xB, yB + 500)
    Fenetre.coords(diametre3, xS, yS ,(2*xB)-xS,(2*yB)-yS)
    Fenetre.coords(diametre4, xB,yB,(2*xS)-xB,(2*yS)-yB)
    Fenetre.coords(diametre5, 2*xB-(yS)+xS-xB,xS+yS-yB,2*xB-((2*yB)-yS)+xS-xB,(2*xB)-xS+yS-yB)
    Fenetre.coords(bulbe, xB - rB, yB - rB, xB + rB, yB + rB)
    w = w_rot_etoile * nb_annees
    xA = xB + distance_etoile_bulbe * math.cos(w)
    yA = yB - distance_etoile_bulbe * math.sin(w)
    Fenetre.coords(etoile, xA - rA, yA - rA, xA + rA, yA + rA)

    w = w_rot_soleil * nb_annees
    xS = xB + distance_soleil_bulbe * math.cos(w)
    yS = yB - distance_soleil_bulbe * math.sin(w)
    Fenetre.coords(soleil, xS - rS, yS - rS, xS + rS, yS + rS)

    w = w_rot_mercure * nb_jours
    xM = xS + distance_mercure_soleil * math.cos(w)
    yM = yS - distance_mercure_soleil * math.sin(w)
    Fenetre.coords(mercure, xM - rM, yM - rM, xM + rM, yM + rM)

    w = w_rot_venus * nb_jours
    xV = xS + distance_venus_soleil * math.cos(w)
    yV = yS - distance_venus_soleil * math.sin(w)
    Fenetre.coords(venus, xV - rV, yV - rV, xV + rV, yV + rV)

    w = w_rot_terre * nb_jours
    xT = xS + distance_terre_soleil * math.cos(w)
    yT = yS - distance_terre_soleil * math.sin(w)
    Fenetre.coords(terre, xT - rT, yT - rT, xT + rT, yT + rT)

    w = w_rot_mars * nb_jours
    xMa = xS + distance_mars_soleil * math.cos(w)
    yMa = yS - distance_mars_soleil * math.sin(w)
    Fenetre.coords(mars, xMa - rMa, yMa - rMa, xMa + rMa, yMa + rMa)

    w = w_rot_jupiter * nb_jours
    xJ = xS + distance_jupiter_soleil * math.cos(w)
    yJ = yS - distance_jupiter_soleil * math.sin(w)
    Fenetre.coords(jupiter, xJ - rJ, yJ - rJ, xJ + rJ, yJ + rJ)

    w = w_rot_saturne * nb_jours
    xSa = xS + distance_saturne_soleil * math.cos(w)
    ySa = yS - distance_saturne_soleil * math.sin(w)
    Fenetre.coords(saturne, xSa - rSa, ySa - rSa, xSa + rSa, ySa + rSa)

    w = w_rot_uranus * nb_jours
    xU = xS + distance_uranus_soleil * math.cos(w)
    yU = yS - distance_uranus_soleil * math.sin(w)
    Fenetre.coords(uranus, xU - rU, yU - rU, xU + rU, yU + rU)

    w = w_rot_neptune * nb_jours
    xN = xS + distance_neptune_soleil * math.cos(w)
    yN = yS - distance_neptune_soleil * math.sin(w)
    Fenetre.coords(neptune, xN - rN, yN - rN, xN + rN, yN + rN)

    xC = xS
    yC = yS
    Fenetre.coords(cercle, xC - rC, yC - rC, xC + rC, yC + rC)

    if compteur == 1:
        if distance(xS, xA, yS, yA)<variable:
            Fenetre.itemconfig(shift_text, text='Blueshift',fill='blue')
            Fenetre.itemconfig(shift,fill='blue')
        elif distance(xS, xA, yS, yA)>variable:
                Fenetre.itemconfig(shift_text, text='Redshift',fill='red')
                Fenetre.itemconfig(shift,fill='red')
        else:
            Fenetre.itemconfig(shift_text, text='No shift',fill='white')
            Fenetre.itemconfig(shift,fill='white')
    if compteur == 0:
        compteur = 1

    if mouvement_etat:
        Fenetre.after(40, mouvement)

def mouvement2():
    global nb_jours, nb_annees, variable, compteur, xA, yA, xB, yB, xS, yS, xM, yM, xV, yV, xT, yT, xMa, yMa, xJ, yJ, xSa, ySa, xU, yU, xN, yN
    nb_annees += 1
    nb_jours += 1
    variable = distance(xS, xA, yS, yA)

    xS = WIDTH/2
    yS = HEIGHT/2
    Fenetre.coords(soleil, xS - rS, yS - rS, xS + rS, yS + rS)

    w = w_rot_bulbe * nb_annees
    xB = xS + distance_soleil_bulbe * math.cos(w)
    yB = yS - distance_soleil_bulbe * math.sin(w)
    Fenetre.coords(bulbe, xB - rB, yB - rB, xB + rB, yB + rB)

    Fenetre.coords(diametre1, xB - 500, yB, xB + 500, yB)
    Fenetre.coords(diametre2, xB, yB - 500, xB, yB + 500)
  
  

    w = w_rot_etoile * nb_annees
    xA = xB + distance_etoile_bulbe * math.cos(w)
    yA = yB - distance_etoile_bulbe * math.sin(w)
    Fenetre.coords(etoile, xA - rA, yA - rA, xA + rA, yA + rA)

    w = w_rot_mercure * nb_jours
    xM = xS + distance_mercure_soleil * math.cos(w)
    yM = yS - distance_mercure_soleil * math.sin(w)
    Fenetre.coords(mercure, xM - rM, yM - rM, xM + rM, yM + rM)

    w = w_rot_venus * nb_jours
    xV = xS + distance_venus_soleil * math.cos(w)
    yV = yS - distance_venus_soleil * math.sin(w)
    Fenetre.coords(venus, xV - rV, yV - rV, xV + rV, yV + rV)

    w = w_rot_terre * nb_jours
    xT = xS + distance_terre_soleil * math.cos(w)
    yT = yS - distance_terre_soleil * math.sin(w)
    Fenetre.coords(terre, xT - rT, yT - rT, xT + rT, yT + rT)

    w = w_rot_mars * nb_jours
    xMa = xS + distance_mars_soleil * math.cos(w)
    yMa = yS - distance_mars_soleil * math.sin(w)
    Fenetre.coords(mars, xMa - rMa, yMa - rMa, xMa + rMa, yMa + rMa)

    w = w_rot_jupiter * nb_jours
    xJ = xS + distance_jupiter_soleil * math.cos(w)
    yJ = yS - distance_jupiter_soleil * math.sin(w)
    Fenetre.coords(jupiter, xJ - rJ, yJ - rJ, xJ + rJ, yJ + rJ)

    w = w_rot_saturne * nb_jours
    xSa = xS + distance_saturne_soleil * math.cos(w)
    ySa = yS - distance_saturne_soleil * math.sin(w)
    Fenetre.coords(saturne, xSa - rSa, ySa - rSa, xSa + rSa, ySa + rSa)

    w = w_rot_uranus * nb_jours
    xU = xS + distance_uranus_soleil * math.cos(w)
    yU = yS - distance_uranus_soleil * math.sin(w)
    Fenetre.coords(uranus, xU - rU, yU - rU, xU + rU, yU + rU)

    w = w_rot_neptune * nb_jours
    xN = xS + distance_neptune_soleil * math.cos(w)
    yN = yS - distance_neptune_soleil * math.sin(w)
    Fenetre.coords(neptune, xN - rN, yN - rN, xN + rN, yN + rN)

    xC = xS
    yC = yS
    Fenetre.coords(cercle, xC - rC, yC - rC, xC + rC, yC + rC)

    if compteur == 1:
        if distance(xS, xA, yS, yA)<variable:
            Fenetre.itemconfig(shift_text, text='Blueshift',fill='blue')
            Fenetre.itemconfig(shift,fill='blue')
        elif distance(xS, xA, yS, yA)>variable:
                Fenetre.itemconfig(shift_text, text='Redshift',fill='red')
                Fenetre.itemconfig(shift,fill='red')
        else:
                    Fenetre.itemconfig(shift_text, text='No shift',fill='white')
                    Fenetre.itemconfig(shift,fill='white')
    if compteur == 0:
        compteur = 1

    if mouvement_etat:
        Fenetre.after(40, mouvement2)

def stopper():
    global mouvement_etat
    mouvement_etat = False


def demarrer():
    global mouvement_etat 
    mouvement_etat = True
    mouvement()
    
    
def demarrer2():
    global mouvement_etat, xS, yS, xB, yB 
    mouvement_etat = True
    xS = WIDTH/2
    yS = HEIGHT/2
    xB = xS + distance_soleil_bulbe
    yB = yS
    mouvement2()
   
    
def reinitialiser():
    global mouvement_etat, xS, yS, nb_annees, nb_jours
    nb_annees = 0
    nb_jours = 0
    mouvement_etat = False
    xB = WIDTH/2
    yB = HEIGHT/2
    Fenetre.coords(diametre1, xB - 500, yB, xB + 500, yB)
    Fenetre.coords(diametre2, xB, yB - 500, xB, yB + 500)
    Fenetre.coords(bulbe, xB - rB, yB - rB, xB + rB, yB + rB)
    xS = xB + distance_soleil_bulbe
    yS = yB
    Fenetre.coords(soleil, xS - rS, yS - rS, xS + rS, yS + rS)
    xM = xS + distance_mercure_soleil
    yM = yS
    Fenetre.coords(mercure, xM - rM, yM - rM, xM + rM, yM + rM)
    xV = xS + distance_venus_soleil
    yV = yS
    Fenetre.coords(venus, xV - rV, yV - rV, xV + rV, yV + rV)
    xT = xS + distance_terre_soleil
    yT = yS
    Fenetre.coords(terre, xT - rT, yT - rT, xT + rT, yT + rT)
    xMa = xS + distance_mars_soleil
    yMa = yS
    Fenetre.coords(mars, xMa - rMa, yMa - rMa, xMa + rMa, yMa + rMa)
    xJ = xS + distance_jupiter_soleil
    yJ = yS
    Fenetre.coords(jupiter, xJ - rJ, yJ - rJ, xJ + rJ, yJ + rJ)
    xSa = xS + distance_saturne_soleil
    ySa = yS
    Fenetre.coords(saturne, xSa - rSa, ySa - rSa, xSa + rSa, ySa + rSa)
    xU = xS + distance_uranus_soleil
    yU = yS
    Fenetre.coords(uranus, xU - rU, yU - rU, xU + rU, yU + rU)
    xN = xS + distance_neptune_soleil
    yN = yS
    Fenetre.coords(neptune, xN - rN, yN - rN, xN + rN, yN + rN)
    xC = xS
    yC = yS
    Fenetre.coords(cercle, xC - rC, yC - rC, xC + rC, yC + rC)
    xA = xB + distance_etoile_bulbe
    yA = yB
    Fenetre.coords(etoile, xA - rA, yA - rA, xA + rA, yA + rA)

def reinitialiser2():
    global mouvement_etat, xS, yS, nb_annees, nb_jours
    nb_annees = 0
    nb_jours = 0
    mouvement_etat = False
    xS = WIDTH/2
    yS = HEIGHT/2
    Fenetre.coords(soleil, xS - rS, yS - rS, xS + rS, yS + rS)
    xB = xS + distance_soleil_bulbe
    yB = yS
    Fenetre.coords(diametre1, xB - 500, yB, xB + 500, yB)
    Fenetre.coords(diametre2, xB, yB - 500, xB, yB + 500)
    Fenetre.coords(bulbe, xB - rB, yB - rB, xB + rB, yB + rB)
    xM = xS + distance_mercure_soleil
    yM = yS
    Fenetre.coords(mercure, xM - rM, yM - rM, xM + rM, yM + rM)
    xV = xS + distance_venus_soleil
    yV = yS
    Fenetre.coords(venus, xV - rV, yV - rV, xV + rV, yV + rV)
    xT = xS + distance_terre_soleil
    yT = yS
    Fenetre.coords(terre, xT - rT, yT - rT, xT + rT, yT + rT)
    xMa = xS + distance_mars_soleil
    yMa = yS
    Fenetre.coords(mars, xMa - rMa, yMa - rMa, xMa + rMa, yMa + rMa)
    xJ = xS + distance_jupiter_soleil
    yJ = yS
    Fenetre.coords(jupiter, xJ - rJ, yJ - rJ, xJ + rJ, yJ + rJ)
    xSa = xS + distance_saturne_soleil
    ySa = yS
    Fenetre.coords(saturne, xSa - rSa, ySa - rSa, xSa + rSa, ySa + rSa)
    xU = xS + distance_uranus_soleil
    yU = yS
    Fenetre.coords(uranus, xU - rU, yU - rU, xU + rU, yU + rU)
    xN = xS + distance_neptune_soleil
    yN = yS
    Fenetre.coords(neptune, xN - rN, yN - rN, xN + rN, yN + rN)
    xC = xS
    yC = yS
    Fenetre.coords(cercle, xC - rC, yC - rC, xC + rC, yC + rC)
    xA = xB + distance_etoile_bulbe
    yA = yB
    Fenetre.coords(etoile, xA - rA, yA - rA, xA + rA, yA + rA)

def quitter():
    MaFenetre.destroy()

def distance(x1, x2, y1, y2):
    return math.sqrt(pow((x2-x1), 2)+pow((y2-y1), 2))
shift_text = Fenetre.create_text((4*WIDTH/5),(4*HEIGHT/5)-50,text='',fill= '')
shift = Fenetre.create_rectangle((4*WIDTH)/5-25,(4*HEIGHT)/5-25,(4*WIDTH)/5+25,(4*HEIGHT)/5+25, fill='')

diametre1 = Fenetre.create_line(
xB-500,
yB,
xB+500,
yB,
fill= 'red',
dash=(4, 4))

diametre2 = Fenetre.create_line(
xB,
yB-500,
xB,
yB+500,
fill= 'red',
dash=(4, 4))

diametre3 = Fenetre.create_line(
xS,
yS,
(2*xB)-xS,
(2*yB)-yS,
fill= 'yellow',
dash=(4, 4))

diametre4 = Fenetre.create_line(
xB,
yB,
(2*xS)-xB,
(2*yS)-yB,
fill= 'yellow',
dash=(4, 4))

diametre5 = Fenetre.create_line(
2*xB-(yS)+xS-xB,
xS+yS-yB,
2*xB-((2*yB)-yS)+xS-xB,
((2*xB)-xS+yS-yB),
fill= 'yellow',
dash=(4, 4))

bulbe = Fenetre.create_oval(
xB - rB,
yB - rB,
xB + rB,
yB + rB,
fill= 'dark blue')

soleil = Fenetre.create_oval(
xS - rS,
yS - rS,
xS + rS,
yS + rS,
fill= 'yellow')

mercure = Fenetre.create_oval(
xM - rM,
yM - rM,
xM + rM,
yM + rM,
fill=c1)

venus = Fenetre.create_oval(
xV - rV,
yV - rV,
xV + rV,
yV + rV,
fill=c2)

terre = Fenetre.create_oval(
xT - rT,
yT - rT,
xT + rT,
yT + rT,
fill=c3)

mars = Fenetre.create_oval(
xMa - rMa,
yMa - rMa,
xMa + rMa,
yMa + rMa,
fill=c4)

jupiter = Fenetre.create_oval(
xJ - rJ,
yJ - rJ,
xJ + rJ,
yJ + rJ,
fill=c5)

saturne = Fenetre.create_oval(
xSa - rSa,
ySa - rSa,
xSa + rSa,
ySa + rSa,
fill=c6)

uranus = Fenetre.create_oval(
xU - rU,
yU - rU,
xU + rU,
yU + rU,
fill=c7)

neptune = Fenetre.create_oval(
xN - rN,
yN - rN,
xN + rN,
yN + rN,
fill=c8)

cercle = Fenetre.create_oval(
xC - rC,
yC - rC,
xC + rC,
yC + rC,
outline=c9)

etoile = Fenetre.create_oval(
xA - rA,
yA - rA,
xA + rA,
yA + rA,
fill=c9)
w = Label(MaFenetre, text= 'Point de vue: centre de la galaxie ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
BoutonDemarrer = Button(MaFenetre, text= ' Démarrer ', command=demarrer)
BoutonDemarrer.pack()
BoutonQuitter = Button(MaFenetre, text= ' Augmenter la vitesse ', command=demarrer)
BoutonQuitter.pack()
BoutonArreter = Button(MaFenetre, text= 'Arrêter ', command=stopper)
BoutonArreter.pack()
BoutonReinit = Button(MaFenetre, text= 'Réinitialiser', command=reinitialiser)
BoutonReinit.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= 'Point de vue: centre du Soleil ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
BoutonStart = Button(MaFenetre, text= 'Démarrer ', command=demarrer2)
BoutonStart.pack()
BoutonAccelerate = Button(MaFenetre, text= 'Augmenter la vitesse ', command=demarrer2)
BoutonAccelerate.pack()
BoutonArreter = Button(MaFenetre, text= 'Arrêter ', command=stopper)
BoutonArreter.pack()
BoutonRestart = Button(MaFenetre, text= 'Réinitialiser ', command=reinitialiser2)
BoutonRestart.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= 'Légende ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= "Démarrer les 2 référentiels en même temps n'est pas recomandé", fg= 'gray')
w.pack()
w = Label(MaFenetre, text= 'Disque bleu: bulbe de la galaxie ', fg= 'blue')
w.pack()
w = Label(MaFenetre, text= 'Tirets rouges: indiquent le diamètre de la galaxie ', fg= 'red')
w.pack()
w = Label(MaFenetre, text= 'Cercle blanc: indique un non-respect de l’échelle de taille ')
w.pack()
w = Label(MaFenetre, text= 'et une échelle différente du temps à l’intérieur ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= 'En jaune, le Soleil, ')
w.pack()
w = Label(MaFenetre, text= 'autour du Soleil les 8 planètes du système solaire ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= 'L’échelle de la taille du bulbe et de la distance du ')
w.pack()
w = Label(MaFenetre, text= 'Soleil au centre de la galaxie est respectée ')
w.pack()
w = Label(MaFenetre, text= 'A l’intérieur du cercle blanc, le temps/vitesse des planètes ')
w.pack()
w = Label(MaFenetre, text= 'est respecté ')
w.pack()
w = Label(MaFenetre, text= ' ')
w.pack()
w = Label(MaFenetre, text= 'En bas à droite, le carré bleu ou rouge indique si l’objet blanc ')
w.pack()
w = Label(MaFenetre, text= 'est observé depuis le Soleil avec un décalage vers le rouge ou le bleu ')
w.pack()
w = Label(MaFenetre, text= 'selon l’effet Doppler-Fizeau ')
w.pack()
w = Label(MaFenetre, text= "pour réduire la vitesse il suffit d'appuyer sur le bouton arrêter puis sur le bouton démarrer")
w.pack()
w = Label(MaFenetre, text= 'Pour plus d’infos, consultez le code ')
w.pack(side= 'bottom' )
BoutonQuitter = Button(MaFenetre, text= 'Quitter ', command=quitter)
BoutonQuitter.pack(side='bottom' )

MaFenetre.mainloop()
```
